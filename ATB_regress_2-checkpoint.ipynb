```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Import dataset
data_df=pd.read_csv('ATB_EC_MM_sample_regression.csv')
```

```python
data_df.head()
```

```python
# define x and y
x=data_df.drop(['BadDebtor','PartyCode'],axis=1).values
y=data_df['BadDebtor'].values
```

```python
print (x)
```

```python
print(y)
```

```python
# split into training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Train the model on the training set
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

```python
# Predict the test set results
y_pred=ml.predict(x_test)
print(y_pred)
```

```python
# test by using known sequence from above to see if predicted value is similar to actual ...ml.predict([[]])
```

```python
# evaluate the model
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
